```python
class NonLinearCov:
    
    def __init__(self, returns, eps = 1e-8):
        
        self.n, self.p = returns.shape
        self.c = self.p / self.n
        self.h_n = self.n ** (-1/3.)
        
        self.sample_cov = returns.cov()
        
        self.D, self.U = np.linalg.eigh(self.sample_cov)
        
        self.eps = eps
        self.is_singular, self.n_singular = self.check_if_is_singular()
        
    def check_if_is_singular(self):
        
        is_singular = sum(self.D <= self.eps) > 0
        n_singular = sum(self.D <= self.eps)
        
        return is_singular, n_singular
        
    def compute_cov(self):
        
        if self.is_singular:
            print('Singular Case')
            
            D_tilde = np.zeros(self.p)
            for i in range(self.p - self.n, self.p):
                D_tilde[i] = self.d_singular(self.D[i])
                
            D_tilde[:self.p - self.n] = self.d_singular_null()
            
            return self.U @ np.diag(D_tilde) @ self.U.T
        
        else:
            print('Non Singular Case')
            D_tilde = np.zeros(self.p)
            for i in range(self.p):
                D_tilde[i] = self.d(self.D[i])

            return self.U @ np.diag(D_tilde) @ self.U.T
        
    def d(self, lambda_ni):
        "Non singular case"
        
        # a / (b + c)
        
        a = lambda_ni
        b = (np.pi * self.c * lambda_ni * self._f_tilde(lambda_ni)) ** 2
        c = (1 - self.c - np.pi * self.c * lambda_ni * self._h_f_tilde(lambda_ni)) ** 2
        
        d = a / (b + c)
        
        return d
    
    def d_singular(self, lambda_ni):
        # a / (b * c)
        
        a = lambda_ni
        b = (np.pi * lambda_ni) ** 2
        c = (self._f_tilde_singular(lambda_ni)** 2  + self._h_f_tilde_singular(lambda_ni) ** 2)
        
        d_singular = a / (b * c)
        
        return d_singular
    
    def d_singular_null(self):
        
        return 1 / (np.pi * ((self.p - self.n) / self.n ) * self._h_f_tilde_singular(0))
    
    def _f_tilde(self, lambda_ni):
        "Non singular case"
        
        f_tilde = 0
        for j in range(self.p):
            # \sum a * b
            a = 3 / (4 * np.sqrt(5) * self._h_nj(j))
            b = np.maximum(1 - 1/5 * ((lambda_ni - self.D[j]) / self._h_nj(j)) ** 2 , 0)
            
            f_tilde = f_tilde + a * b
            
        f_tilde = f_tilde / self.p    
        
        return f_tilde
    
    def _h_f_tilde(self, lambda_ni):
        "Non singular case"
        
        h_f_tilde = 0
        for j in range(self.p):
            # \sum a + b * c * d
            
            a = - 3 * (lambda_ni - self.D[j]) / (10 * np.pi * self._h_nj(j) ** 2)
            b = 3 / (4 * np.sqrt(5) * np.pi * self._h_nj(j))
            c = 1 - 1/5 * ((lambda_ni - self.D[j]) / self._h_nj(j)) ** 2
            d = np.log(np.abs((np.sqrt(5) * self._h_nj(j) - lambda_ni + self.D[j]) / (np.sqrt(5) * self._h_nj(j) + lambda_ni - self.D[j])))
            
            h_f_tilde = h_f_tilde + a + b * c * d 
        
        h_f_tilde = h_f_tilde / self.p
        
        return h_f_tilde
    
    def _h_nj(self, j):
        "Non singular case"
        
        return self.D[j] * self.h_n
    
    def _f_tilde_singular(self, lambda_ni):
        
        f_tilde_singular = 0
        for j in range(self.p - self.n, self.p):
            # \sum a * b
            a = 3 / (4 * np.sqrt(5) * self.D[j] * self.h_n)
            b = np.maximum((1 - (1/5) * ((lambda_ni - self.D[j]) / (self._h_nj(j) * self.h_n) ** 2)) , 0)
            
            f_tilde_singular = f_tilde_singular + a * b
            
        f_tilde_singular = f_tilde_singular / self.n
        
        return f_tilde_singular
            
    
    def _h_f_tilde_singular(self, lambda_ni):
        
        is_null = lambda_ni == 0
        if is_null:
            # a + b * c * \sum
            a = 3 / (10 * self.h_n ** 2)
            b = ((3)/(4 * np.sqrt(5) * self.h_n) * (1 - 1 / (5 * self.h_n ** 2)))
            c = np.log((1 + np.sqrt(5) * self.h_n) / (1 - np.sqrt(5) * self.h_n))
            
            h_f_zero = 0
            for j in range(self.p - self.n, self.p):
                h_f_zero = h_f_zero + (1 / self.D[j])
            
            h_f_zero = h_f_zero / (np.pi * self.n)
            
            return a + b * c * h_f_zero
        
        else:
            h_f_tilde_singular = 0
            for j in range(self.p - self.n, self.p):
                # \sum a + b * c * d

                a = - 3 * (lambda_ni - self.D[j]) / (10 * np.pi * (self.D[j] ** 2) * (self.h_n ** 2))
                b = 3 / (4 * np.sqrt(5) * self.D[j] * self.h_n)
                c = 1 - (1/5) * ((lambda_ni - self.D[j]) / (self._h_nj(j) * self.h_n) ** 2)
                d = np.log(np.abs((np.sqrt(5) * self._h_nj(j) * self.h_n - lambda_ni + self.D[j]) / (np.sqrt(5) * self._h_nj(j) * self.h_n + lambda_ni - self.D[j])))

                h_f_tilde_singular = h_f_tilde_singular +  a + b * c * d

            h_f_tilde_singular = h_f_tilde_singular / self.n
        
        return h_f_tilde_singular
```

```python
class DirectKernel(object):

    def __init__(self, X):
        self.X = X
        self.n = None
        self.p = None
        self.sample = None
        self.eigenvalues = None
        self.eigenvectors = None
        self.L = None
        self.h = None

    def pav(self, y):
        """
        PAV uses the pair adjacent violators method to produce a monotonic
        smoothing of y
        translated from matlab by Sean Collins (2006) as part of the EMAP toolbox
        """
        y = np.asarray(y)
        assert y.ndim == 1
        n_samples = len(y)
        v = y.copy()
        lvls = np.arange(n_samples)
        lvlsets = np.c_[lvls, lvls]
        flag = 1
        while flag:
            deriv = np.diff(v)
            if np.all(deriv >= 0):
                break

            viol = np.where(deriv < 0)[0]
            start = lvlsets[viol[0], 0]
            last = lvlsets[viol[0] + 1, 1]
            s = 0
            n = last - start + 1
            for i in range(start, last + 1):
                s += v[i]

            val = s / n
            for i in range(start, last + 1):
                v[i] = val
                lvlsets[i, 0] = start
                lvlsets[i, 1] = last
        return v

    def estimate_cov_matrix(self):

        # extract sample eigenvalues sorted in ascending order and eigenvectors
        self.n, self.p = self.X.shape
        self.sample = (self.X.transpose() @ self.X) / self.n
        self.eigenvalues, self.eigenvectors = np.linalg.eig(self.sample)
        isort = np.argsort(self.eigenvalues, axis=-1)
        self.eigenvalues.sort()
        self.eigenvectors = self.eigenvectors[:, isort]

        # compute direct kernel estimator
        self.eigenvalues = self.eigenvalues[max(1, self.p - self.n + 1) - 1:self.p]
        self.L = np.repeat(self.eigenvalues, min(self.n, self.p), axis=0).reshape(self.eigenvalues.shape[0], min(self.n, self.p))
        self.h = self.n ** (-0.35)
        component_00 = 4*(self.L.T**2)*self.h**2 - (self.L - self.L.T)**2
        component_0 = np.maximum(np.zeros((component_00.shape[1], component_00.shape[1])), component_00)
        component_a = np.sqrt(component_0)
        component_b = 2*np.pi*(self.L.T**2)*self.h**2
        ftilda = np.mean(component_a / component_b, axis=1)

        com_1 = np.sign(self.L - self.L.T)
        com_2_1 = (self.L - self.L.T)**2 - 4*self.L.T**2*self.h**2
        com_2 = np.maximum(np.zeros((com_2_1.shape[1], com_2_1.shape[1])), com_2_1)
        com_3_1 = np.sqrt(com_2)
        com_3_2 = com_1 * com_3_1
        com_3 = com_3_2 - self.L + self.L.T
        com_4 = 2*np.pi*self.L.T**2*self.h**2
        com_5 = com_3 / com_4
        Hftilda = np.mean(com_5, axis=1)

        if self.p <= self.n:
            com_0 = (np.pi*(self.p/self.n)*self.eigenvalues*ftilda)**2
            com_1 = (1 - (self.p / self.n) - np.pi * (self.p / self.n) * self.eigenvalues * Hftilda) ** 2
            com_2 = com_0 + com_1
            dtilde = self.eigenvalues / com_2
        else:
            Hftilda0 = (1-np.sqrt(max(1-4*self.h**2, 0))) / (2*np.pi*self.n*self.h**2)*np.mean(1/self.eigenvalues)
            dtilde0 = 1/(np.pi*((self.p-self.n)/self.n)*Hftilda0)
            dtilde1 = self.eigenvalues/np.pi**2*self.eigenvalues**2*(ftilda**2+Hftilda**2)
            dtilde = np.hstack((dtilde0*np.ones((self.p-self.n, 1)).reshape(self.p-self.n,), dtilde1))

        dhat = self.pav(dtilde)
        d_matrix = np.diag(dhat)
        sigmahat = self.eigenvectors.dot(d_matrix).dot(self.eigenvectors.T)
        return sigmahat
```
